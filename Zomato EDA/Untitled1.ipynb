```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
 final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
#pie chart 
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observation: Zomato maximum records or transactions are from India

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## Observation : -- When rating is between 4.5 to 4.9 ----> food is excellent
## when ratings is 4.0 to 4.4 -----> food is very good
# when ratings is between 3.5 to 3.9 ----> food is good
# when ratings is between 2.5 to 3.4 ----> food is Average
# when ratings is below 2.4 ----> food is Poor
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color')
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['grey','red','orange','yellow','green','green'])
```

```python
#Observation 
# 1.Not rated count is very high
# 2.Maximum no.of ratings are between 2.7 to 4.1
```

```python
# count plot
sns.countplot(x="Rating color",data=ratings,hue='Rating color',palette=['grey','red','orange','yellow','green','green'])
```
